## Is there any correlation between different values in the dataset?

```python
import matplotlib.pyplot as plt
```

```python
# Filter only numeric or float columns
numeric_columns = df.select_dtypes(include=['number'])  # Select numeric columns

# Compute correlation matrix
correlation_matrix = numeric_columns.corr()
```

```python
f = plt.figure(figsize=(12, 8))
plt.matshow(correlation_matrix, fignum=f.number)
plt.xticks(range(numeric_columns.shape[1]), numeric_columns.columns, fontsize=14, rotation=45)
plt.yticks(range(numeric_columns.shape[1]), numeric_columns.columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.show()
```

```python
# Calculate the count of null values for each column
null_counts = df.isnull().sum()
```

```python
null_counts
```

```python
# Plot the counts using a bar plot
plt.figure(figsize=(10, 6))
null_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Columns')
plt.ylabel('Count of Null Values')
plt.title('Count of Null Values in Each Column')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculate the percentage of null values for each column
total_rows = len(df)
null_percentages = (df.isnull().sum() / total_rows) * 100

# Step 2: Plot the percentages using a bar plot
plt.figure(figsize=(10, 6))
null_percentages.plot(kind='bar', color='skyblue')
plt.xlabel('Columns')
plt.ylabel('Percentage of Null Values')
plt.title('Percentage of Null Values in Each Column')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Apply function

### for this exercise, we will look at a dataset where we have Apple's stock price information

```python
df = pd.read_csv("https://raw.githubusercontent.com/imanojkumar/Datasets/master/AAPL.csv")
```

```python
df.head()
```

 - To create a date column, we will create a function `'date_to_year'` and use `apply` function to apply it

```python
def date_to_year(row):
    return int(row['Date'].split('-')[0])   # split on '-' and take out the first value from splitted data
```

```python
df.apply(date_to_year, axis=1)  # apply function on the column 
```

Above returned the 'years', to save it into dataframe we do:

```python
df['Year'] = df.apply(date_to_year, axis=1)
```

# Data Cleaning in Pandas

```python
# reading data from an excel file on github
df = pd.read_excel("https://github.com/imanojkumar/Datasets/raw/master/Customer_Call_List.xlsx")
df.head()
```

```python
df.shape
```

```python
# unique values (please do not do it, when data is really large), to get an idea
df['Last_Name'].unique()
```

### As you can notice that there are some '/' or '...' in the data and the data is not clean... so we must clean the data before any of the analysis

```python
df
```

### Also we notice that there are duplicate rows (such as 19 and 20), so we must remove duplicates

```python
# drop duplicates
df = df.drop_duplicates()
```

```python
df.tail()
```

 - we removed the duplicate row

```python
df.columns
```

### Moreover, there's a column 'Not_Useful_Column' which should be removed from the dataframe as it is of no use for our analysis

```python
# let's do this
df = df.drop(columns = "Not_Useful_Column")
df.head()
```

 - so we removed it

```python
#df["Last_Name"] = df["Last_Name"].str.lstrip("...")
#df["Last_Name"] = df["Last_Name"].str.lstrip("/")
#df["Last_Name"] = df["Last_Name"].str.rstrip("_")

df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

 - as we can see, Last_Name columns is now has no numbers or dot or dash or backslash

### as we can see that Phone_Number contains `-` and `|` and we must get rid of all such things 

```python
# although we can remove one by one the garbage from Phone_Number

df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')

df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')

df["Phone_Number"] = df["Phone_Number"].str.replace('n/a--','')

df["Phone_Number"] = df["Phone_Number"].str.replace('N/a--','')

df["Phone_Number"] = df["Phone_Number"].str.replace('N/a','')

df["Phone_Number"] = df["Phone_Number"].str.replace('n/a','')

df.head()
```

```python
# df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')

# df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

# df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))

# df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
# split the address into street address, state and zip code

df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',', expand=True)
df
```

```python
# let's replace Yes and No with a Y and N

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')

df
```

```python
# replace NAs in the dataset
df=df.fillna('')
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('N/a','')
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')

df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')

df
```
